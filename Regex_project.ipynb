```python
import urllib.request
import re
hand = urllib.request.urlopen('https://en.wikipedia.org/wiki/Elon_Musk')


def ascii_histogram(seq) -> None: #create a histogram from a list of numbers
    counted = count_elements(seq) #generate frequency table (dictionary)
    for key in sorted(counted):
        plus_str = '+' * counted[key]
        print(f'{key} {plus_str}') #prints out asterisk
def count_elements(seq): #creates a dictionary
    hist = dict()
    for date in seq:
        hist[date]=hist.get(date,0)+ 1
    return hist

years = []
for line in hand:
    line = line.decode().strip()
    years_found = re.findall('\s[0-9]{4}',line)
    if years_found  != []:
        years = years + years_found
    # your code here: fill up the list dates using regex and dates.append()
ascii_histogram(years)





```
